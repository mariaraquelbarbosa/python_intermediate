```python
# Ex.12 - Case básico

aceito = "PAGAMENTO ACEITO!"
fiado = "FIADO!"
recusado = "QUANTIA INSUFICIENTE"

limite = - 50

print("Nome:" + ' ' * 11 + "Status:" + ' ' * 18 + "Saldo final:")

def poder_consumidor (nome, money, pg):
    saldo = float(money - pg)
    
    if saldo > 0:
        status = aceito
    else:
        if saldo >= limite:
            status = fiado
        else:
            status = recusado
            saldo = money
    
    print(f"{nome:15} {status:24} $ {saldo:.2f}")

##
poder_consumidor("Thomas", 420, 500)
poder_consumidor("Ed", 1240, 1290)
poder_consumidor("Alice", 480, 500)
poder_consumidor("Edu", 500, 320)
```

###        1.1.4 Retornando valores

```python
# Ex.4 - Case

Mont_inicial = 9000

# Função para automatizar a conta usando os parâmetros de cada investimento)
def Mont(tx_juros, tempo):
    Mf = Mont_inicial * ((1 + tx_juros) ** (tempo))
    
    return Mf

mont_final1 = Mont(0.018, 19)
maior = mont_final1
name = 1

mont_final2 = Mont(0.02, 14)
if mont_final2 > maior:
    maior = mont_final2
    name = 2

# Resposta
print(f"O maior montante final pertence ao investimento {name}, sendo de R${maior:.2f}")
```

        2.1.2 Difereça entre variável global e local

```python
# Ex.2 Case básico - 
# Não se pode chamar variável local no escopo global
const = 0.0175


def cal(speed, weight):
    calo = (speed * weight * const)
    return calo 

h = cal(8, 65)
cal_final = h
name = "Henry Ford"

k = cal(12, 58)
if k < cal_final:
    cal_final = k
    name = "Kiichiro Toyoda"

print("A pessoa com o menor gasto calórico é {}, sendo de {}/min".format(name, round(cal_final, 3)))
```

```python
# Ex.2 Case básico - A variável local não altera o valor da variável global
const = 0.0175

def cal(speed, weight):
    #calo = (speed * weight * const)
    const = (speed * weight * const)
    return const

h = cal(8, 65)
cal_final = h
name = "Henry Ford"

k = cal(12, 58)
if k < cal_final:
    cal_final = k
    name = "Kiichiro Toyoda"

print("A pessoa com o menor gasto calórico é {}, sendo de {}/min".format(name, round(cal_final, 3)))
```

```python
# A variável global não pode ser alterada dentro da função
const = 0.0175
cal_final = 9999
name = ''

def cal(speed, weight):
    calo = (speed * weight * const)
    return calo


def replace(current_cal, current_name):
    if current_cal < cal_final:
        cal_final = current_cal
        name = current_name
    return cal_final


h = cal(8, 65)
replace(h, "Henry Ford")

k = cal(12, 5)
replace(k, "Kiichiro Toyoda")


print(f"A pessoa com o menor gasto calórico é {name}, sendo de {cal_final:.2f}/min")
```

```python
# #global pode ser usado para esse fim

const = 0.0175
current_cal = 0
cal_final = 9999
name_final = ''


def replace(speed, weight, name):
    global cal_final
    global name_final

    current_cal = (speed * weight * const)

    if current_cal < cal_final:
        cal_final = current_cal
        name_final = name

    print(name_final)
    print(cal_final)

replace(8, 65, "Henry Ford")
replace(12, 58, "Kiichiro Toyoda")
replace(11, 60, "Bernd Elberskirch")

print(f"A pessoa com o menor gasto calórico é {name_final}, sendo de {cal_final:.2f}/min")
```

- ### 2.2 Funções aninhadas



```python
# Estrutura:

def outer( ... ):
    """ ... """
    x = ...
    
    def inner( ... ):
        """ ... """
        y = x ** 2
    return ...
```

```python
def mod2plus5(x1, x2, x3):
    """Returns the remainder plus 5 of three values."""
    
    def inner(x):
        """Returns the remainder plus 5 of a value."""
        return x % 2 + 5
    
    return (inner(x1), inner(x2), inner(x3))

print(mod2plus5(1, 2, 3))
```

###        2.2.1 Retornando Funções

```python
def raise_val(n):
    """Return the inner function."""
    
    def inner(x):
        """Raise x to the power of n."""
        raised = x ** n
        return raised
    
    return inner

square = raise_val(2)
cube = raise_val(3)
print(square(2), cube(4))
```

###        2.2.2 Usando "nonLocal"

- Serve para se acessar o valor de variável na função "superior"

- LEGB rule, primeiro o programa procura a variável na função local, depois nas funções acima, depois nas variáveis globais e depois nas funçõs built ins do python

```python
#####


def outer():
    """Prints the value of n."""
    n = 1
    
    def inner():
        #nonlocal n
    
        n = 2
        print(n)
        
    inner()
    print(n)
    
outer()
```

```python
''' Para o parâmetro default é só usar um “=” e o valor padrão na criação da função,
nesse caso se for para usar o valor padrão do argumento não é necessário preencher o argumento'''

def power(number, pow=float(1)):
    """Raise number to the power of pow."""
    new_value = number ** pow
    return new_value

power(9, 2)
```

###         2.3.2 Argumentos flexíveis "*args"

- Os nomes "*args" e "**kwargs" são usados apenas por convenção
- É usado para se passar uma lista de argumentos de forma dinâmica sem palavras-chave
- Os argumentos passados são armazenados em uma tupla de nome "args"

```python
# Exemplo 1:

def add_all(*args):
    """Sum all values in *args together."""
    
    # Initialize sum
    sum_all = 0
    
    # Accumulate the sum
    for num in args:
        sum_all += num
        
    return sum_all
```

```python
add_all()
```

```python
add_all(1)
```

```python
add_all(7, 99)
```

```python
add_all(999, 10, 0.75, 2775)
```

```python
# Exemplo 2:

def func_args(*args):
    print(f'tipo: {type(args)} conteúdo: {args}')
    for arg in args:
        print(f'tipo: {type(arg)} conteúdo: {arg}')


func_args(1, 'A', {'valor': 10})

# Fonte código: https://medium.com/luizalabs/fun%C3%A7%C3%B5es-em-python-entendendo-par%C3%A2metros-argumentos-args-e-kwargs-4291b1f817f6
```

###        2.3.3 Argumentos flexíveis "**kwargs" (Keyword Arguments)
- Argumentos precedidos por identificadores
- Basicamente permite passar um dicionário com inúmeras chaves para a função

```python
# Exemplo 1:

def print_all(**kwargs):
    """Print out key-value pairs in **kwargs."""
    
    # Print out the key-value pairs
    for key, value in kwargs.items():
        print(key + ':' + value)
              
print_all(name="dumbledore", job="headmaster")
```

```python
# Exemplo 2:

def func_kwargs(**kwargs):
    print(f'tipo: {type(kwargs)} contuedo: {kwargs}')
    for key, value in kwargs.items():
        print(f'atributo: {key}, valor: {value}')

func_kwargs(nome='James', sobrenome='Bond')

# Fonte código: https://medium.com/luizalabs/fun%C3%A7%C3%B5es-em-python-entendendo-par%C3%A2metros-argumentos-args-e-kwargs-4291b1f817f6
```
